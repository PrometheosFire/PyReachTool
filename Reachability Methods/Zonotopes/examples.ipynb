```python
import numpy as np
from zonotope import Zonotope
```

Create Zonotope

```python
# Create a simple zonotope
center = np.array([0, 0])
generators = np.array([[1, 0], [0, 1], [1, 1]])
zonotope = Zonotope(generators.T, center)

print(f"Number of doubles: {zonotope.nbDoubles()}" )
zonotope.volume()
```

Plot Zonotope

```python
zonotope.plot()
```

Propagate phase

```python
A = np.random.rand(2, 2)
L = np.random.rand(2, 2)
B = np.random.rand(2, 2)

system = {}
system['A'] = A
system['L'] = L
system['B'] = B
u = np.array([1, 1]).reshape(-1, 1)
d = zonotope #assume disturbance is equal to the state for illustration

z1 = zonotope.propagate(system, u, d)

z1.plot()
```

Update Phase

```python
noise = z1 # assume noise is equal to the state for illustration

C = np.eye(2)
N = np.eye(2)
D = np.zeros((2, 2))

sys2 = {}
sys2['C'] = C
sys2['N'] = N
sys2['D'] = D

u = np.array([1, 1]).reshape(-1, 1)
y = np.array([2, 2]).reshape(-1, 1)

z2 = z1.update(sys2, u, y, noise)

z2.plot()
```
